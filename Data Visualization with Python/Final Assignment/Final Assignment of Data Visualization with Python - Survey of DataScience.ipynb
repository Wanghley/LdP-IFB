```
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```
df_survey = pd.read_csv('df_survey.csv',index_col=0)
df_survey.head()
```

```
df_survey
```

```
df_survey.sort_values('Very interested',axis=0,ascending=False,inplace=True)
df_survey
```

```
total_responses = 2233
```

```
ax = df_survey.plot(kind='bar',figsize=(20,8),title='Result of DataScience\'s survey')
fig = ax.figure

for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x(), i.get_height()+12, \
            str(round((i.get_height()/total_responses)*100, 2))+'%',
                color='black')

plt.legend(loc='upper right')
plt.ylabel('no. of responses')
fig.savefig('barchart01.png')
plt.show()
```
