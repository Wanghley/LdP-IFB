### Pie Charts

- Porcentagem de participação na imigração canadense de 1980-2013

Etapa 1: coletar dados.

Usaremos o método * pandas * `groupby` para resumir os dados de imigração por` Continent`. O processo geral de `groupby` envolve os seguintes passos:

1. ** Dividir: ** Dividir os dados em grupos com base em alguns critérios.
2. ** Aplicar: ** Aplicar uma função a cada grupo independentemente:
        .sum()
        .count()
        .mean() 
        .std() 
        .aggregate()
        .apply()
        .etc..
3. ** Combinar: ** Combinar os resultados em uma estrutura de dados.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Images/Mod3Fig4SplitApplyCombine.png" height=400 align="center">

### Box plots

- Gráficos de caixas <a id="8"> </a>

Um `box plot` é uma maneira de representar estatisticamente a * distribuição * dos dados através de cinco dimensões principais:

- **Mínimo:** Menor número no conjunto de dados.
- **Primeiro quartil:** Número do meio entre o `mínimo 'e a` mediana`.
- **Segundo quartil (mediana):** Número do meio do conjunto de dados (classificado).
- **Terceiro quartil:** Número médio entre "mediana" e "máxima".
- **Máximo:** Número mais alto no conjunto de dados.

#### Subplots
Muitas vezes, podemos querer plotar vários gráficos na mesma figura. Por exemplo, podemos querer fazer uma comparação lado a lado do gráfico de caixa com o gráfico de linha da imigração da China e da Índia.

Para visualizar várias plotagens juntas, podemos criar uma **`figure`** (tela geral) e dividi-la em **`subplots`**, cada uma contendo uma plotagem. Com **subplots**, geralmente trabalhamos com a **Artist layer** em vez da **Script layer**.

Sintaxe típica é: <br>
```python
    fig = plt.figure() # cria figure
    ax = fig.add_subplot(nrows, ncols, plot_number) # cria subplots
```
Onde
- `nrows` e `ncols` são usados para nocionalmente splittar a figura em (`nrows` \* `ncols`) sub-axes,  
- `plot_number` é usado para identificar o subplot particular que essa função é criar dentro da grade nacional. `plot_number` começa em 1, incrementa primeiro as linhas e tem no máximo `nrows` * `ncols` como mostrado abaixo.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Images/Mod3Fig5Subplots_V2.png" width=500 align="center">

````
fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

df_CI.plot(kind='box', color='blue',vert=False,figsize=(20,5),ax=ax0,showfliers=False)
ax0.set_title('Box plot de imigração da China e ÍÍndia 1980 - 2013')
ax0.set_xlabel('Nº de imigrantes')
ax0.set_ylabel('Países')

df_CI.plot(kind='line',figsize=(20,5),ax=ax1)
ax1.set_title('Line plot de imigração da China e ÍÍndia 1980 - 2013')
ax1.set_ylabel('Nº de imigrantes')
ax1.set_xlabel('Anos')
````

**Pergunta:** Crie um gráfico de caixa para visualizar a distribuição dos 15 principais países (com base na imigração total) agrupados pelas *décadas* `1980s`,` 1990s` e `2000s`.

````
df_can.sort_values('total',axis=0,ascending=False,inplace=True)
df_top15 =  df_can.head(15)

years_80 = list(range(1980,1990))
years_90 = list(range(1990,2000))
years_2000 = list(range(2000,2010))

#Divide o dataframe original df_top15 para criar uma série para cada década
df_top15_80 = df_top15.loc[:,years_80].sum(axis=1)
df_top15_90 = df_top15.loc[:,years_90].sum(axis=1)
df_top15_2 = df_top15.loc[:,years_2000].sum(axis=1)

#Merge dos 3 dataframes em um único dataframe
new_df = pd.DataFrame({'1980s':df_top15_80,'1990s':df_top15_90,'2000s':df_top15_2})

new_df.head()
````

````
new_df.describe()
````

````
#Plotagem
new_df.plot(kind='box',figsize=(15,8))
plt.title("Distribuição dos TOP 15 países em imigração pro Canadá",y=1.05)
plt.show()
````

Observe como a plotagem da caixa difere da tabela de resumo criada. O gráfico da caixa verifica os dados e identifica os **valores discrepantes**(outliers). Para ser um outlier, o valor dos dados deve ser: <br>
* maior que Q3 em pelo menos 1,5 vezes o intervalo interquartil (IQR) ou
* menor que Q1 em pelo menos 1,5 vezes o IQR.

Vejamos a década de 2000 como um exemplo: <br>
* Q1 (25%) = 36.101,5 <br>
* Q3 (75%) = 105.505,5 <br>
* IQR = Q3 - Q1 = 69.404 <br>

Usando a definição de outlier, qualquer valor maior que Q3 em 1,5 vezes o IQR será sinalizado como outlier.

Outlier> 105.505,5 + (1,5 * 69.404) <br>
Outlier> 209.611,5

Etapa 2: Crie os pesos normalizados.

Existem vários métodos de normalização na estatística, cada um com seu próprio uso. Nesse caso, usaremos [recurso de escala](https://en.wikipedia.org/wiki/Feature_scaling) para trazer todos os valores para o intervalo [0,1]. A fórmula geral é:

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Images/Mod3Fig3FeatureScaling.png" align = "center">

onde *`X`* é um valor original, *`X'`* é o valor normalizado. A fórmula define o valor máximo no conjunto de dados como 1 e define o valor mínimo como 0. O restante dos pontos de dados é dimensionado para um valor entre 0-1 em conformidade.

````
#Normalizando os dados do Brasil
norm_brazil = (df_canada_t['Brazil']-df_canada_t['Brazil'].min())/(df_canada_t['Brazil'].max()-df_canada_t['Brazil'].min())

#Normalizando os dados da Argentina
norm_argentina = (df_canada_t['Argentina'] - df_canada_t['Argentina'].min()) / (df_canada_t['Argentina'].max() - df_canada_t['Argentina'].min())
````

Agora, seria muito ineficiente repetir essas sete etapas toda vez que desejamos criar um gráfico de waffles. Então, vamos combinar todas as sete etapas em uma função chamada * create_waffle_chart *. Esta função usaria os seguintes parâmetros como entrada:

> 1. **categories**: categorias ou classes exclusivas no dataframe.
> 2. **values**: valores correspondentes a categorias ou classes.
> 3. **height**: altura definida da tabela de waffles.
> 4. **width**: largura definida do gráfico de waffles.
> 5. **colormap**: classe Colormap
> 6. **value_sign**: Para tornar nossa função mais generalizável, adicionaremos esse parâmetro para tratar de sinais que possam estar associados a um valor como%, $ e assim por diante. **value_sign** possui um valor padrão de string vazia.

````
def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):

    # Proporção entre os valores
    total_values = sum(values)
    category_proportions = [(float(value) / total_values) for value in values]

    # computa o número de azulejos
    total_num_tiles = width * height # total
    print ('Total number of tiles is', total_num_tiles)
    
    # no. de azulejo por categoria
    tiles_per_category = [round(proportion * total_num_tiles) for proportion in category_proportions]

    # printa o no. de azulejos por categoria

    # inicializa o waffle como uma matriz preenchida por zeros
    waffle_chart = np.zeros((height, width))

    category_index = 0
    tile_index = 0

    # preenche o waffle chart
    for col in range(width):
        for row in range(height):
            tile_index += 1

            # Se o no. de azulejos preenchidos é igual ao valor alocado 
            if tile_index > sum(tiles_per_category[0:category_index]):
                # ...vá para a próxima categoria
                category_index += 1       
            
            # preenche a matriz
            waffle_chart[row, col] = category_index
    
    # instancia novo objeto de figure
    fig = plt.figure()

    # use matshow para mostrar o gráfico
    colormap = plt.cm.coolwarm
    plt.matshow(waffle_chart, cmap=colormap)
    plt.colorbar()

    # obtém o eixo(axis)
    ax = plt.gca()

    # define minor ticks
    ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
    ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
    # adiciona gridlines baseados nos minor ticks
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

    plt.xticks([])
    plt.yticks([])

    # computa soma cumulativa de cada categoria para alinhar a cor à legenda a ser apresentada
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum[len(values_cumsum) - 1]

    # creia legenda
    legend_handles = []
    for i, category in enumerate(categories):
        if value_sign == '%':
            label_str = category + ' (' + str(values[i]) + value_sign + ')'
        else:
            label_str = category + ' (' + value_sign + str(values[i]) + ')'
            
        color_val = colormap(float(values_cumsum[i])/total_values)
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    # adiciona legenda ao gráfico
    plt.legend(
        handles=legend_handles,
        loc='lower center', 
        ncol=len(categories),
        bbox_to_anchor=(0., -0.2, 0.95, .1)
    )
````

- Exemplo de uso da função:

````
width = 50
height=10

categories = df_dns.index.values
values = df_dns['total']

colormap = plt.cm.coolwarm

create_waffle_chart(categories,values,height,width,colormap)
````

### Mapas Choropleth
Um mapa `Choropleth` é um mapa temático no qual as áreas são sombreadas ou padronizadas na proporção da medida da variável estatística exibida no mapa, como densidade populacional ou renda per capita. O mapa choropleth fornece uma maneira fácil de visualizar como uma medida varia em uma área geográfica ou mostra o nível de variabilidade dentro de uma região. Abaixo está um mapa dos EUA 'Choropleth', representando a população por quilômetro quadrado por estado.

````
world_map.choropleth(
    geo_data=world_geo,
    data=df_canada2,
    columns=['country','total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    legend_name='Immigration to Canada'
)

world_map
````

````
#Altera para deixar a legenda mais forte
threshold_scale = np.linspace(df_canada2['total'].min(),
                              df_canada2['total'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # altera de array para lista
threshold_scale[-1] = threshold_scale[-1] + 1 # ter certeza que é maior que o maior valor do df

world_map = folium.Map(location=[0, 0], zoom_start=1.25, tiles='Mapbox Bright')
world_map.choropleth(
    geo_data=world_geo,
    data=df_canada2,
    columns=['country', 'total'],
    key_on='feature.properties.name',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada',
    reset=True
)
world_map
````
